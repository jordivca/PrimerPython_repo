```python
#Data de ventas
# En este laboratorio crearemos un ETL que genere un reporte mensual de las ventas y algunas vistas transformadas sobre el dataset
path = r"D:\ANALYTIC_PYTHON\EntornoPython\ETL DATASET - 2\Dataset_ ventas"

df = pd.read_excel(path +"\\Sample - Superstore.xlsx")
```

```python
#examinemos el dataset
df.head(3)
```

```python
#veamos la calidad de la data del dataframe
df.isna().sum()
```

```python
#si el profit no tiene nulos se podria estimar las ventas a partir de un % de la ganancia
promedio_porcentaje_ganancias = df['Profit'].sum()/df['Sales'].sum()
```

```python
# cañculemos para los casos donde no se tengan el valor de la venta , lo remplazamos el valor profit/0.12

df['Sales'] = np.where(df['Sales'].isna(), df['Profit']/promedio_porcentaje_ganancias, df['Sales'])
```

```python
#Sobre los datos que son categoricos se realizara un cambio del nulo por el estring "sin informacion"

df['Product ID']  = df['Product ID'].fillna("Sin Info")     
df['Category']    = df['Category'].fillna("Sin Info")  
df['Sub-Category']   = df['Sub-Category'].fillna("Sin Info")            
df['Product Name']   = df['Product Name'].fillna("Sin Info")         

#df['Product ID']  = np.where( df['Product ID'].isna(),"Sin Info",df['Product ID'])
```

```python
#veamos si aun quedan valores nulos
df.isna().sum()
```

```python
df.dtypes
```

```python
#construyamos el reporte mensualizado por mes y año y producto

df['Mes'] = df['Order Date'].dt.month
df['Año'] = df['Order Date'].dt.year
```

```python
#df.groupby(['Mes','Año']).agg({"Sales":"sum","Profit":"sum"})
df.groupby(['Mes','Año']).agg({"Sales":"sum","Profit":"sum"}).reset_index()
```

```python
#construimos el reporte de año y mes
reporte_mes_año = df[df["Año"].isin([2014,2016])].groupby(['Mes','Año']).agg({"Sales":"sum","Profit":"sum"}).reset_index()
```

```python
#Miremos el reporte
reporte_mes_año.head(5)
```

```python
# Creemos algunas variables
# calculemos el casto de fabricación como el 50% de las ventas
# Calculamos el IGV como el 18% de las ventas
reporte_mes_año["Costo_ventas"] = reporte_mes_año["Sales"] * 0.5
reporte_mes_año["IGV"] = reporte_mes_año["Sales"] * 0.18
```

```python
#Miremos el reporte
reporte_mes_año.head(5)
```

```python
#Generemos el reporte por regiones y sus manager
```

```python
#Maestro de regiones y managet
df_region = pd.read_excel(path +"\\Sample - Superstore.xlsx", sheet_name = "People")
```

```python
#Muestra de la nueva base de datos
df_region
```

```python
#realizaremos un merge
df_union = df.merge(df_region, on=["Region"], how = "left")
```

```python
# Reporte por gerente zonales
reporte_gerente_zonal = df_union.groupby(["Person"]).agg({"Sales":"sum","Profit":"sum"}).reset_index()
reporte_gerente_zonal
```

```python
# Creemos algunas variables
# calculemos el casto de fabricación como el 50% de las ventas
# Calculamos el IGV como el 18% de las ventas
reporte_gerente_zonal["Costo_ventas"] = reporte_gerente_zonal["Sales"] * 0.5
reporte_gerente_zonal["IGV"] = reporte_gerente_zonal["Sales"] * 0.18
```

```python
#Calculemos el ranking
reporte_gerente_zonal["Rank"] = reporte_gerente_zonal["Sales"].rank(ascending=False)
```

```python
reporte_gerente_zonal.head()
```

```python
#crear reporte de productos
reporte_productos = df.groupby(["Category"]).agg({"Sales":"sum","Profit":"sum"}).reset_index()
reporte_productos.sort_values("Sales",ascending=False)
```

```python
#reporte clientes
reporte_clientes = df.groupby(["Customer Name"]).agg({"Sales":"sum","Profit":"sum"}).reset_index()



# Creemos algunas variables
# calculemos el casto de fabricación como el 50% de las ventas
# Calculamos el IGV como el 18% de las ventas
reporte_clientes["Costo_ventas"] = reporte_clientes["Sales"] * 0.5
reporte_clientes["IGV"] = reporte_clientes["Sales"] * 0.18


#Calculemos el ranking
reporte_clientes["Rank"] = reporte_clientes["Sales"].rank(ascending=False)
reporte_clientes.sort_values('Rank').head()
```

```python
#Finalmente realizaremos la carga de ambos reportes
pword = "jcelisa27"
name ="sa"
host = 'LAPTOP-J0JK4H2E\DEVELOPER'
database ='jardineria'
#Carga desde SQL
import pyodbc
cnxn = pyodbc.connect(driver='{SQL Server Native Client 11.0}', 
                      host= host, 
                      database= database,
                      trusted_connection='yes',
                      user=name, password=pword)

#HABILITAMOS Y DESHABILITAMOS CUANDO SE CREA POR PRIMERA VEZ PARA LUEGO ELIMINARLA Y VOLVERLA A CREAR

cur=cnxn.cursor()
cur.execute("TRUNCATE TABLE REPORTE_MENSUAL")  
cur.commit()
cur.execute("TRUNCATE TABLE REPORTE_GERENTES")  
cur.commit()
cur.execute("TRUNCATE TABLE REPORTE_CLIENTES")  
cur.commit()

cnxn.close()
engine = create_engine("mssql+pyodbc://" + name + ":" + pword + "@" + host +"/" + database + "?driver=ODBC+Driver+17+for+SQL+Server")

reporte_mes_año.to_sql('REPORTE_MENSUAL', con = engine, if_exists = 'append', chunksize = 10000)
reporte_gerente_zonal.to_sql('REPORTE_GERENTES', con = engine, if_exists = 'append', chunksize = 10000)
reporte_clientes.to_sql('REPORTE_CLIENTES', con = engine, if_exists = 'append', chunksize = 10000)
```
