```python
import pandas as pd
dfTitanic = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
dfTitanic
```

```python
#Pregunta 1: Identificar los valores nulos del dataset Titanic
dfTitanic.isnull().sum()
```

```python
#Pregunta 2:
#Calcular la cantidad de sobrevivientes del titanic que sean hombres, 
#que hayan estado en tercera clase y que sean mayores de 20 años.
df_Varones_Tc_20 = dfTitanic[(dfTitanic['Sex'] == 'male') & (dfTitanic['Pclass'] == 3) & (dfTitanic['Age'] > 20)]
df_Varones_Tc_20['Sex'].value_counts()
```

```python
#Pregunta 3: Crear una nueva columna que indique si eran menor de edad (usar ‘Si’ y ‘No’).
import numpy as np
dfTitanic['MenorEdad'] = np.where(dfTitanic['Age'] >= 18, 'NO', 'Si')
```

```python
#Pregunta 4: Mostrar el histograma de los sobrevivientes en función de su edad.
df_Sobreviviente = dfTitanic[(dfTitanic['Survived'] == 1)]
df_Sobreviviente['Age'].hist()
```
